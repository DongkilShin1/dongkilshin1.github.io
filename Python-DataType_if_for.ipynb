```python
# list member function  # 리스트 멤버 함수
# append - 마지막에 추가, extend - 병합, insert - 중간에도 추가 가능, remove, pop, index, in, 
# append
a1 = [1,2,3,4,5]
a2 = [6,7,8,9,10]
a3 = [11, 12]
a1.append(10)
print(a1)
# extend()
a1.append(a2) # append와 extend 비교
print(a1)
print(a1[6])  # [6,7,8,9,10]
a1.extend(a2)
print(a1)
print(a1[7]) # 6 
a1 += a3   # a1.extend(a3) 와 동일 # a1 = a1 + a3
print(a1)
# insert()
a3.insert(1, 40) # (index, object:추가할값) # shift + tab
a3.insert(2, 12)
print(a3)
# remove
a3.remove(12)   # remove first occurence of value : 처음 값만 지움
print(a3)
# d = a3.remove(12)
# print(d)          # 결과는 None : 즉 아무것도 없음, remove는 반환 안 함 # pop은 반환함
# pop    # 지우고 반환도 함
a3.pop()    # index 없으면 마지막것 지움
print(a3)
d = a3.pop(0)   # index 있으면 index에 해당하는 값 지움
print(d)
# index
print(a2.index(7))   # 값이 위치한 index 번호 반환
# in
b = 7
c = b in a2   # a2 안에 b가 있는지 확인(true or false)
print(c)
# sort, sorted    # sort - 멤버함수, 메소드; sorted - 내장함수  # sort 는 변수 내용 바꾸고, sorted는 변수 내용 안바꿈
e = [2,1,3,5,6,4,9,8]
e.sort()
print(e)
e.sort(reverse=True)
print(e)
d = sorted(e)
print(d)
```

```python
# Tuple : multiple return 가능
a = 100, 200  # 괄호 생략 가능
print(type(a))
a, b = 100, 200  # tuple unpacking  # 여러개 변수에 선언과 동시에 값 대입
print(a, b)
a, b = b, a     # tuple unpacking; 원래는 temp = a; a = b; b = temp
print(a, b)
```
